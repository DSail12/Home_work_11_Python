Даны две функции: 
f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241
Требуется:
Найти точки пересечения
Построить графики функций в одной системе координат
Построить график общей функции

```python
from sympy import *
x = Symbol('x')
fx = x**3 - 50*x
gx = -1*x**4 + 88*x**2 - 241
fgx = fx - gx
```

Корни

```python
roots = real_roots(fgx)
roots[0].evalf(5)
```

Словарь с X и Y

```python
cross_dict = {}
c = 1
for z in roots:
    cross_dict[c] = {'x': round(N(z), 5), 'y': round(N(-1*z**4 + 88*z**2 - 241), 5)}
    c += 1
```

```python
cross_dict
```

Вывод координат точек пересечения

```python
print(*[(cross_dict[i]['x'],cross_dict[i]['y']) for i in cross_dict.keys()], sep=', ')
```

Вывод графиков функций в одной системе координат

```python
print(*[(cross_dict[i]['x'], cross_dict[i]['y']) for i in cross_dict.keys()], sep=', ')
```
